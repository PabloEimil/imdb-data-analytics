#### **1.1. Â¿QuÃ© gÃ©neros tienen mejores valoraciones medias?**

```python
# Numero de votos por categoria

df1= df.groupby("titleType")["numVotes"].sum().sort_values(ascending=False)

plt.bar(df1.index, df1.values )
plt.show()
```

```python
#Media de votos por categorÃ­a

df2= df.groupby("titleType")["averageRating"].mean().sort_values(ascending=False)

plt.bar(df2.index, df2.values)
```

```python
#RelaciÃ³n votos y valoracion

plt.scatter(df["numVotes"], df["averageRating"], alpha=0.1)
plt.xscale("log")
plt.xlabel("NÃºmero de votos (log)")
plt.ylabel("ValoraciÃ³n media")
plt.title("RelaciÃ³n entre nÃºmero de votos y valoraciÃ³n")
plt.show()
```
