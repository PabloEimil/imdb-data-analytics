En esta gráfica se muestra la relación entre la duración de las películas y su valoración promedio, medida a través del número promedio de votos.

Se observa que las películas con más votos tienden a tener valoraciones entre **7 y 9** y duraciones de **120 a 220** minutos.

También se aprecia que los votos están bastante polarizados: hay muchas películas con valoraciones bajas (menos de 3) y muchas con valoraciones altas (más de 7), mientras que las valoraciones intermedias (3-7) acumulan menos votos.

En cuanto a la duración ideal de una película, aunque las más largas pueden tener más votos si alcanzan una alta valoración, existe un riesgo significativo: las películas largas con valoraciones bajas apenas reciben votos. Por ello, una estrategia más segura sería mantener la duración entre **100 y 140 minutos**, ligeramente por encima de la media, donde los votos se distribuyen de manera más equilibrada y no tan polarizada según la valoración.

Por último, esta gráfica resalta un hallazgo clave del estudio: las películas con valoraciones altas, entre 7 y 10, reciben muchos más votos que aquellas con notas inferiores.

-   ### **<u>Conclusión</u>**

El análisis revela que no existe un género que destaque simultáneamente en cantidad de títulos, votos y valoración, aunque Acción, Aventura, Crimen y Thriller logran un equilibrio interesante. Drama, Comedia y Documental dominan en número de títulos, mientras que Documental, Biografía e Historia destacan por valoración media.

La duración influye de forma moderada: el rango óptimo parece estar entre 100 y 140 minutos, evitando producciones muy largas con baja valoración, que suelen recibir pocos votos. Entre los géneros tradicionales, las duraciones son bastante similares (80-100 minutos), salvo Acción, ligeramente más larga, y Western, Documental y Animación, algo más cortos.

**En conjunto, los proyectos más prometedores serían aquellos que combinen géneros populares con buena valoración y duración en el rango óptimo, evitando los extremos y los géneros atípicos de corta duración.**
